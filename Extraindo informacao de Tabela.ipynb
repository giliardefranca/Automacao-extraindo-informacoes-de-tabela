```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

#abrindo navegador
navegador = webdriver.Chrome()
navegador.get("https://pbdatatrader.com.br/jogosdodia/")

#entrando no primeiro iframe
iframe = navegador.find_element(By.TAG_NAME, 'iframe')
navegador.switch_to.frame(iframe)

#entrando no segundo iframe
iframe = navegador.find_element(By.TAG_NAME, 'iframe')
navegador.switch_to.frame(iframe)

#esperando o  elemento
while len(navegador.find_elements(By.CLASS_NAME, 'innerContainer')) == 0:
    time.sleep(1)

    
#coletando o conteudo da tabela  
tabela = navegador.find_element(By.CLASS_NAME, 'innerContainer').text
tabela = tabela.split('\n')

#separando as colunas e linhas
lista_coluna = tabela[:10]
lista_coteudo = tabela[10:]

#criando um dicionario
dicionario = {}

for coluna in lista_coluna:
    dicionario[coluna] = []
    for i in range(20):
        dicionario[coluna].append(lista_coteudo[i])
    lista_coteudo = lista_coteudo[20:]
    
#tranformando em um DataFrame
tabela_df = pd.DataFrame().from_dict(dicionario)


#fazendo scroll na tabela e coletando as informaçoes e fazendo todo processo anterior
for i in range(1000):
    scroll = i * 50
    navegador.execute_script(f"document.getElementsByClassName('bodyCells')[0].scroll(0, {scroll})")
    
    tabela = navegador.find_element(By.CLASS_NAME, 'innerContainer').text

    tabela = tabela.split('\n')

    lista_coluna = tabela[:10]
    lista_conteudo = tabela[10:]

    dicionario = {}

    for coluna in lista_coluna:
        dicionario[coluna] = []
        for i in range(20):
            dicionario[coluna].append(lista_conteudo[i])
        lista_conteudo = lista_conteudo[:20]
    
    #criando uma tabela auxiliar e adicionando na tabela pricinpal
    tabela_aux = pd.DataFrame().from_dict(dicionario)
    tabela_df = tabela_df.append(tabela_aux)

#removendo informaçoes duplicada    
tabela_df = tabela_df.drop_duplicates(ignore_index=False)

display(tabela_df)
```
